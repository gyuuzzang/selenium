```python
all_data = []
for i in range(2, 12):
    tmp = '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)
    sel_ele = driver.find_elements_by_xpath(tmp)

    column_data = []
    for one in sel_ele:
        column_data.append(one.text)

    print( len(sel_ele) )
    all_data.append(column_data)
    print(column_data)
    print()

all_data
```

```python
import pandas as pd

dict_dat = { '국가':all_data[0],
             '확진자':all_data[1],
             '치료중':all_data[2],
             '위중증':all_data[3],
             '사망자':all_data[4],
             '완치' :all_data[5],
             '치명(%)':all_data[6],
             '완치(%)':all_data[7],
             '발생률':all_data[8],
             '인구수':all_data[9]
           }

dat = pd.DataFrame(dict_dat)
dat
```

```python
dat['확진자수'] = dat['확진자'].str.split('\n').str[0]
dat['확진자 1일'] = dat['확진자'].str.split('\n').str[1]

dat['치료중수'] = dat['치료중'].str.split('\n').str[0]
dat['치료중 1일'] = dat['치료중'].str.split('\n').str[1]

dat['위중증수'] = dat['위중증'].str.split('\n').str[0]
dat['위중증 1일'] = dat['위중증'].str.split('\n').str[1]

dat['사망자수'] = dat['사망자'].str.split('\n').str[0]
dat['사망자 1일'] = dat['사망자'].str.split('\n').str[1]

dat['완치수'] = dat['완치'].str.split('\n').str[0]
dat['완치 1일'] = dat['완치'].str.split('\n').str[1]


dat = dat.drop( [ '확진자', '치료중','위중증','사망자', '완치'], axis=1)
dat.head()
```

```python
dat['치명(%)'] = dat['치명(%)'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat['완치(%)'] = dat['완치(%)'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat['발생률'] = dat['발생률'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['인구수'] = dat['인구수'].str.replace(pat=r'[,]', repl=r'', regex=True)

dat['확진자수'] = dat['확진자수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['확진자 1일'] = dat['확진자 1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)

dat['치료중수'] = dat['치료중수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['치료중 1일'] = dat['치료중 1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)

dat['위중증수'] = dat['위중증수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['위중증 1일'] = dat['위중증 1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)

dat['사망자수'] = dat['사망자수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['사망자 1일'] = dat['사망자 1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)

dat['완치수'] = dat['완치수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['완치 1일'] = dat['완치 1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)

dat.head()
```

```python
dat = dat.astype({'확진자수': 'float32'})
```
