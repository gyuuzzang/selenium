```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10, size=10)
print(len(x), type(x),x)

y = np.random.randint(80,100, size=10)
print(len(y), type(y),y)

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar() # colormap 을 포함할 수 있다.
plt.show()
```

```python
score = Series( [1000, 14000, 3000, 3000, 1000] )
print(score)
print("자료형 확인 : ", type(score))
```

```python
## Series 인덱스 확인
print(score.index)
# 인덱스를 리스트 자료형으로 변경 후, 확인하기
print(list(score.index))
## Series 값 확인
print(score.values)
## Series 값 자료형 확인
print(score.dtype) 
```

```python
### 인덱스(index) 속성 이용
score = Series( [1000, 14000, 3000],
 index = ['2019-05-01', '2019-05-02', '2019-05-03'] )
print(score) 
```

```python
print(type(score['2019-05-01']))
print(score['2019-05-01'])
```

```python
print(type(score['2019-05-02':'2019-05-03']))
print(score['2019-05-02':'2019-05-03']) #series는 index값 표함된다
```

```python
for idx in score.index:
 print(idx)
```

```python
for value in score.values:
 print(value) 
```

```python
from pandas import DataFrame
team_score = { "toto":[1500,3000,5000,7000,5500],
 "apple":[4000,5000,6000,5500,4500],
 "gildong":[2000,2500,3000,4000,3000],
 "catanddog":[7000,5000,3000,5000,4000]}
team_df = DataFrame(team_score)
team_df
```

```python
date = ['19-05-01','19-05-02', '19-05-03', '19-05-04', '19-05-05']
team_df = DataFrame(team_score,
 columns=['catanddog', 'toto', 'apple', 'gildong'],
 index=date)
team_df
```

```python
print(team_df.loc[:,:]) #전체 열 행 선택
```

```python
print(team_df.loc[:,"toto"]) #toto열 선택
```

```python
print(team_df.loc[:,"toto":"apple"])
```

```python
print(team_df.loc['19-05-01',:]) #행 선택
```
