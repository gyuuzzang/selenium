```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4,5], 'o') # '-' 선 : 기본값/ y값임
plt.show()
```

```python
import numpy as np
ypoints = np.array([3, 3, 1, 20, 10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
x = range(0, 100)
y = [ v * v for v in x ]
plt.plot(x, y, 'ro')
plt.show()
# ro'에서 'r'은 red를 의미하고, 'o'는 그래프의 마커 모양을 의미한다.
```

```python
y = np.array([3, 3, 1, 20, 10])
plt.plot(y , marker = 'o', ms=15, mec = 'r')
plt.show()
```

```python
plt.plot(y , marker = 'o', ms=15, mfc = 'r')
plt.show()
# 마커안에 색상 :mfc
```

```python
fig = plt.figure(figsize=(10,6)) # 기본: 6.4/4.8
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(1,3,1) # 1행 3열 의 것중에 첫번째
ax2 = fig.add_subplot(1,3,2) # 1행 3열 의 것중에 두번째
ax3 = fig.add_subplot(1,3,3) # 1행 3열 의 것중에 두번째
plt.show()
```

```python
### plot, bar 그래프 그리기
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
x=range(0,100)
y=[v*v for v in x]
ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
x = np.arange(10)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.plot(x, 2*x, label='y = %ix' % 2)
plt.legend()
plt.show()
# 범례 그리기: label을 쓰고 legend
```
