```python
df.iplot(kind='scatter', x='A',y='B',mode='markers',
 size=20, color='#3f9142', symbol=15)
```

```python
df.iplot(kind='scatter', x='A',y='C',mode='markers+text',
 size=20, color='green', symbol=5,
        title="A와 C의 산점도 그래프", xTitle="A", yTitle="C")
```

```python
df.iplot(kind='scatter', x='A',y='C',mode='markers+text',
 size=20, color='yellow', symbol='star-open',
        title="A와 C의 산점도 그래프", xTitle="A", yTitle="C")
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
iris.info()
```

```python
iris.columns
```

```python
iris['sepal_length'].iplot(kind='line')
```

```python
iris.iplot(kind='bar',x='species',y='sepal_length')
```

```python
iris.iplot(kind='bar',x='sepal_width',y='sepal_length',
           title="width와 length 상관관계"
           ,xTitle='sepal_width',yTitle='sepal_length')
```

```python
* 상관관계가 없어 보인다.
* 넓이가 3(중간값)에 가까워질 수록 길이가 길다.
```

```python
iplot.subplot(2,1,1)
iris.iplot(kind='bar',x='species',y='sepal_length')

iplot.subplot(2,1,2)
iris.iplot(kind='bar',x='species',y='sepal_length')
```

```python
iris.mean()
```

```python
iris['mean']=iris.iloc[:,0:4].mean(axis=1)
iris
```

```python
iris.iplot(kind='bar', x='species',
           xTitle='Species', yTitle = 'all')
```

```python
import plotly.express as px

# iris 데이터 불러오기
print(px.data.iris.__doc__)
px.data.iris().head()
```

```python
iris = px.data.iris()
print(type(iris))
fig=px.scatter(iris, x="sepal_width",y="sepal_length")
fig
```

```python
fig=px.scatter(iris, x="sepal_width",y="sepal_length",color='species')
fig
```

```python
fig=px.scatter(iris, x="sepal_width",y="sepal_length",color='species',
              marginal_y='violin',
              marginal_x="box")
fig
```
