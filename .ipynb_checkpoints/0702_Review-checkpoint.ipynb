```python
from pandas import DataFrame

team_score = { "toto":[1500,3000,5000,7000,5500],
              "apple":[4000,5000,6000,5500,4500],
              "gildong":[2000,2500,3000,4000,3000],
              "catanddog":[7000,5000,3000,5000,4000]}

team_df = DataFrame(team_score)
team_df
```

```python
team_df.columns
```

```python
team_df.columns = ['catanddog', 'gildong', 'apple', 'toto']
team_df.columns
```

```python
team_df = DataFrame(team_score, columns=['catanddog',
                                        'gildong',
                                        'apple',
                                        'toto'])
team_df
```

```python
team_df[ ['gildong', 'apple']  ]
```

```python
team_df.loc[ :, 'gildong':'toto'  ]
```

```python
team_df
```

```python
team_df.iloc[ : , 1:4 ]
```

```python
### 
print( team_df.sum() )
print( team_df.mean() )
print( team_df.median() )
print( team_df.std() )
print( team_df.var() )
```

```python
### 
print( team_df.sum() )
print( team_df.mean() )
print( team_df.median() )
print( team_df.std() )
print( team_df.var() )
```

```python
team_df.toto.sort_values(ascending=False)
```

```python
team_df['rowsum'] = team_df.sum(axis=1)
team_df
```

```python
### 행의 합 중에 17000이상인 것만 보고 싶다.
team_df[ team_df['rowsum'] >= 17000 ]
```

```python
### 행의 합 중에 17000이상인 것만 보고 싶다.
team_df.loc[ team_df['rowsum'] >= 17000 , : ]
```

```python
team_df = team_df.drop( 'rowsum', axis=1  )
```

```python
grouped = tips.groupby('sex')
grouped.count()  # mean(), sum(), std(), max(), min()...
```

```python
grouped = tips.groupby(['sex', 'smoker'])
grouped.count()  # mean(), sum(), std(), max(), min()...
```

### 실습 1-2

### 실습 1-3
 * 남성의 데이터만 따로 만들어보자.(day가 Sun만 뽑기) male_sun
 * 여성의 데이터만 따로 만들어보자.(day가 Sun만 뽑기) female_sum


```python
tips.groupby(['sex', 'day']).max()
```

```python
grouped = tips.groupby(['sex', 'day'])
grouped.get_group('Female', 'Sun')
```

```python
grouped = tips.groupby(['sex', 'day'])
grouped.count()
```

```python
male_sun = grouped.get_group( ( 'Male', 'Sun' ) )
male_sun
```

```python
# * sex, day를 그룹화시켜서 최대값을 출력해 보자.
tips.groupby(['sex', 'day']).max()
```

```python
#  sex, day를 그룹화시켜서 최대값의 tip과 size 열만 출력해 보자.
tips.groupby(['sex', 'day']).max()[ ['tip', 'size']  ]
```
