```python
sel_country = driver.find_elements_by_xpath('//*[@id="country-table"]/div/div/table/tbody/tr/td[2]')
len(sel_country)
```

```python
all_data = []
for i in range(2, 11):
    tmp = '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)
    sel_ele = driver.find_elements_by_xpath(tmp)

    column_data = []
    for one in sel_ele:
        column_data.append(one.text)

    print( len(sel_ele) )
    all_data.append(column_data)
    print(column_data)
    print()

all_data
```

```python
import pandas as pd
dict_dat = { "Country":all_data[0],
 "Confirmed":all_data[1],
 "Active Case":all_data[2],
 "Deaths":all_data[3],
 "Recovered":all_data[4],
 "Fatality(%)":all_data[5],
 "Recovery(%)":all_data[6],
 "Incidence":all_data[7],
 "Population":all_data[8]
 }
dat = pd.DataFrame(dict_dat)
dat
```

```python
dat['Confirmed_all'] = dat['Confirmed'].str.split('\n').str[0]
dat['Confirmed 1day'] = dat['Confirmed'].str.split('\n').str[1]
dat['Active Case_all'] = dat['Active Case'].str.split('\n').str[0]
dat['Active Case 1day'] = dat['Active Case'].str.split('\n').str[1]
dat['Deaths_all'] = dat['Deaths'].str.split('\n').str[0]
dat['Deaths 1day'] = dat['Deaths'].str.split('\n').str[1]
dat['Recovered_all'] = dat['Deaths'].str.split('\n').str[0]
dat['Recovered 1day'] = dat['Deaths'].str.split('\n').str[1]
dat = dat.drop( [ 'Confirmed', 'Active Case', 'Deaths','Recovered'], axis=1)
dat.head()
```
