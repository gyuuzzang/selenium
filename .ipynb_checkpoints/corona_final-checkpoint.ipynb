# 가설) 1일 확진자수와 확진자수의 상관관계가 있을 거 같다.
* 1일 확진자수를 기준으로 상위, 하위 10개 국가 추출

```python
import pandas as pd
dict_dat = { "국가":all_data[0],
  "확진자":all_data[1]
 }
dat = pd.DataFrame(dict_dat)

dat['국가'] = dat['국가'].str[:-2]
dat['확진자수'] = dat['확진자'].str.split('\n').str[0]
dat['확진자1일'] = dat['확진자'].str.split('\n').str[1]
dat = dat.drop( [ '확진자'], axis=1)

dat['확진자수'] = dat['확진자수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['확진자1일'] = dat['확진자1일'].str.replace(pat=r'[,() +]', repl=r'', regex=True)

dat = dat.astype({'확진자1일': 'float32'})
dat = dat.astype({'확진자수': 'float32'})
dat = dat.sort_values("확진자1일",axis=0, ascending=True)
dat.head(10)
```

```python
dat_low10 = dat.iloc[0:10 , :] #data_ten = dat.head(10)
dat_low10 
```

```python
import pandas as pd
dict_dat = { "국가":all_data[0],
  "확진자":all_data[1]
 }
dat = pd.DataFrame(dict_dat)

dat['국가'] = dat['국가'].str[:-2]
dat['확진자수'] = dat['확진자'].str.split('\n').str[0]
dat['확진자1일'] = dat['확진자'].str.split('\n').str[1]
dat = dat.drop( [ '확진자'], axis=1)

dat['확진자수'] = dat['확진자수'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['확진자1일'] = dat['확진자1일'].str.replace(pat=r'[,() +]', repl=r'', regex=True)

dat = dat.astype({'확진자1일': 'float32'})
dat = dat.astype({'확진자수': 'float32'})
dat = dat.sort_values("확진자1일",axis=0, ascending=False)
dat.head(10)
```

```python
dat_top10 = dat.iloc[0:10 , :] #data_ten = dat.head(10)
dat_top10 
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import platform

if platform.system()=='Windows':
    plt.rc('font',family= 'Malgun Gothic')
fig = plt.figure(figsize=(31,14))

plt.subplot(2,2,2)
sns.set_color_codes('pastel')
sns.barplot(data=dat_top10 , x='국가', y='확진자수',label='total', alpha=0.5)

plt.subplot(2,2,1)
sns.set_color_codes('muted')
sns.barplot(data=dat_top10 , x='국가', y='확진자1일',label='day')

plt.subplot(2,2,4)
sns.set_color_codes('pastel')
sns.barplot(data=dat_low10 , x='국가', y='확진자수',label='total', alpha=0.5)

plt.subplot(2,2,3)
sns.set_color_codes('muted')
sns.barplot(data=dat_low10 , x='국가', y='확진자1일',label='day')

plt.show()
```

# 결론
* 상위 10국은 브라질, 인도 등 1일 확진자가 많은 국가들은 전체 확진자수 또한 상당하다는 것을 알 수 있다.
    * 미국만 예외적인 양상을 보인다. 
    * 이유: 접종 횟수 3.2억, 접종완료1.51억으로 접종 완료 인구별 비율 45.9% 가 백신을 접종하였기 때문에 1일 확진자 수가 적은 것을 알 수 있다.

*하위 10국은 1일 확진자수가 적은 만틈 전체 확진자수도 적은 편임을 알 수 있다.
    *알바니아가 예외적인 양상을 보인다.
    
*가설분석: 일반적으로 확진자 1일수와 확진자수는 비례적인 관계를 가지고 있음을 알 수 있다.

## 개선할 사항
* 바 그래프에 수치 표기를 하면 좋겠다.
* 확진자 1일과 확진자수가 한 그래프에 표시하는 방법 고안할 필요가 있다.
    * 확진자 수가 확진자 1일에 비해 너무 수치가 크기 때문에 각각 그래프를 만들었다.
    * 확진자수 수치를 만단위로 나누어 수치를 변환하는 방법을 시도해 볼 생각이다.

## 1일 확진자수와 확진자 수의 상관관계를 명확하게 표시하고자 jointplot 이용
* 조인트 플롯을 사용하면 두개의 수치형 변수 간의 관계를 연구할 수 있다.

```python
sns.jointplot(y="확진자수", x="확진자1일", data=dat_top10, kind="kde")
plt.suptitle("확진자 1일 상위 10개 국가", y=1.02)

plt.show()
```

```python
sns.jointplot(y="확진자수", x="확진자1일", data=dat_low10, kind="kde")
plt.suptitle("확진자 1일 하위 10개 국가", y=1.02)

plt.show()
```

### 1일 확진자 상위 국가가 하위국가보다는 뚜렷한 상관관계를 보이고 있음을 알 수 있다.

### 개선할 사항
* jointplot에 대한 공부 더 필요 : 박스 양 옆이 나타내는 그래프가 무엇인지 모름.
