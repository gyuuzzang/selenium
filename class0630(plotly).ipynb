```python
fig = make_subplots(rows=1, cols=2, shared_xaxes=True)


gap_1952 = gapminder.query("year==1952")
gap_2007 = gapminder.query("year==2007")

gap_1952_line = go.Scatter(x=gap_1952["pop"],
 y=gapminder["lifeExp"],
                          mode='markers', name='1952')

gap_2007_line = go.Scatter(x=gap_2007["pop"],
 y=gap_2007["lifeExp"],
 mode='markers', name='2007')

fig.add_trace(gap_1952_line, row=1, col=1)
fig.add_trace(gap_2007_line, row=1, col=2)

fig
```

```python
### 실습 5-1 (추가) -px활용
 * 2행, 2열(왼쪽-1952년, 오른쪽 2007년)
 * 1행 - 인구, 기대수명
 * 2행 - 인구, 1인당 국민소득
```

```python
fig = make_subplots(rows=2, cols=2, shared_xaxes=True)


gap_1952 = gapminder.query("year==1952")
gap_2007 = gapminder.query("year==2007")

gap1= go.Scatter(x=gap_1952["pop"],
 y=gapminder["lifeExp"],
   mode='markers',name='1952 기대수명')

gap2 = go.Scatter(x=gap_2007["pop"],
 y=gap_2007["lifeExp"],
 mode='markers', name='2007 기대수명')

gap3= go.Scatter(x=gap_1952["pop"],
 y=gapminder["gdpPercap"],
   mode='markers',name='1952 국민소득')

gap4 = go.Scatter(x=gap_2007["pop"],
 y=gap_2007["gdpPercap"],
 mode='markers', name='2007 국민소득')

fig.add_trace(gap1, row=1, col=1)
fig.add_trace(gap2, row=1, col=2)
fig.add_trace(gap3, row=2, col=1)
fig.add_trace(gap4, row=2, col=2)

ayout = {
 "xaxis": {
 "title": "인구",
 },
 "yaxis": {
 "title": "기대수명"
 },

 "xaxis2": {
 "title": "인구",
 },
 "yaxis2": {
 "title": "인구"
 },
    
 "xaxis3": {
 "title": "인구",
 },
 "yaxis3": {
 "title": "인당 국민소득"
 },
 "xaxis4": {
 "title": "인구",
 },
 "yaxis4": {
 "title": "인당 국민소득"
 }

}
fig.update_layout(layout)

fig
```
