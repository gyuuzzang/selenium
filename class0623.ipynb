```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,14,100) # 0부터 14까지 동일한 간격의 값 100개 만들기
print(type(x), len(x))

y1 = np.sin(x)
y2 = np.sin(x+0.5) * 2
y3 = np.sin(x+1.0) * 4

plt.figure(figsize=(10,6))   # 그림의 크기
plt.plot(x, y1)
```

```python
plt.plot(x,y1, x, y2, x, y3)
```

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='2*sin(x)')
plt.plot(x, y3, label='3*sin(x)')

plt.legend()
```
